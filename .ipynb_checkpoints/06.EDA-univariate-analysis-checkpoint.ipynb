```python
#visualize single variables(Univariate Analysis)
#shape,center,spread
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#magic command to display - plots inside notebook
%matplotlib inline

#default style for plots 
sns.set_style('whitegrid')

#load the data
data = {
    'timestamp': pd.to_datetime([
        '2025-07-30 08:30:00', '2025-07-30 09:15:00', '2025-07-30 09:45:00',
        '2025-07-30 12:00:00', '2025-07-30 13:30:00', '2025-07-30 17:00:00',
        '2025-07-30 18:30:00', '2025-07-30 19:00:00', '2025-07-30 21:00:00',
        '2025-07-30 22:30:00'
    ]),
    'cab_type': ['Mini', 'Sedan', 'Mini', 'Sedan', 'Mini', 'SUV', 'Sedan', 'Mini', 'Sedan', 'Mini'],
    'distance_km': [5.5, 12.1, 3.2, 8.0, 4.5, 6.7, 15.3, 4.1, 9.8, 7.6],
    'temperature_celsius': [28.5, 29.0, 29.2, 32.0, 31.5, 29.8, 27.5, np.nan, 26.0, 25.5],
    'weather_condition': ['Clear', 'Clear', 'Clear', 'Clear', 'Clear', 'Rainy', 'Rainy', 'Rainy', 'Clear', 'Foggy'],
    'demand_multiplier': [1.0, 1.5, 1.2, 1.0, 1.0, 2.0, 2.5, 2.2, 1.2, 1.8]
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)
df
```

```python
#set size of plot
plt.figure(figsize=(10,6))

#create the histogram
sns.histplot(df['demand_multiplier'], kde=True , bins=30) #Kernel Density Estimate is a smooth line on the plot to see data distribution more clearly.

#add titles for clarity
plt.title('Distribution of Demand Multiplier')
plt.xlabel('Demand Multiplier')
plt.ylabel('Frequency')

#show the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))

#create the box plot
sns.boxplot(x=df['distance_km'])

plt.title('Box plot of ride distance (km)')
plt.xlabel('Distance (km)')

plt.show()
```

```python
#task for the day 
plt.figure(figsize=(10,6))

sns.histplot(df['temperature_celsius'], kde=True, bins=30)

plt.title('A Temeprature Histogram')
plt.xlabel('Temperature')
plt.ylabel('Frequency')

plt.show()


#It shows how your temperature readings are distributed. For example, you can see clusters of data around 26°C, 28-30°C, and 32°C.
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x=df['temperature_celsius'])

plt.title('A Temeprature Box plot')
plt.xlabel('Degree Celsius')

plt.show()



```

```python
#The plot shows that while temperatures ranged from 25.5°C to 32°C, the majority of the time (the middle 50%), the temperature was between a narrow range of 28°C and 30°C.
```
