```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
sns.set_style('whitegrid')

data = {
    'timestamp': pd.to_datetime([
        '2025-07-30 08:30:00', '2025-07-30 09:15:00', '2025-07-30 09:45:00',
        '2025-07-30 12:00:00', '2025-07-30 13:30:00', '2025-07-30 17:00:00',
        '2025-07-30 18:30:00', '2025-07-30 19:00:00', '2025-07-30 21:00:00',
        '2025-07-30 22:30:00'
    ]),
    'cab_type': ['Mini', 'Sedan', 'Mini', 'Sedan', 'Mini', 'SUV', 'Sedan', 'Mini', 'Sedan', 'Mini'],
    'distance_km': [5.5, 12.1, 3.2, 8.0, 4.5, 6.7, 15.3, 4.1, 9.8, 7.6],
    'temperature_celsius': [28.5, 29.0, 29.2, 32.0, 31.5, 29.8, 27.5, np.nan, 26.0, 25.5],
    'weather_condition': ['Clear', 'Clear', 'Clear', 'Clear', 'Clear', 'Rainy', 'Rainy', 'Rainy', 'Clear', 'Foggy'],
    'demand_multiplier': [1.0, 1.5, 1.2, 1.0, 1.0, 2.0, 2.5, 2.2, 1.2, 1.8]
}

df = pd.DataFrame(data)
df
```

```python
#scatter plot
#to see if there's a connection between ride distance & demand multiplier

plt.figure(figsize=(10,6))

sns.scatterplot(x='distance_km' , y='demand_multiplier', data=df)

plt.title('Demand Multiplier vs Ride Distance')
plt.xlabel('Distance(km)')
plt.ylabel('Demand Multiplier')
plt.show()
```

```python
#to select only numerical columns
numerical_df = df.select_dtypes(include=['float64','int64'])

#calculataing correlation matrix    
correlation_matrix = numerical_df.corr()

#visualize the matrix using heatmap
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True, cmap = 'coolwarm' , fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()                                                                                     
```

```python
#temperature celsius vs demand multiplier scatter plot
plt.figure(figsize=(10,6))

sns.scatterplot(x='temperature_celsius' , y='demand_multiplier', data=df)

plt.title('Demand Multiplier vs temperature celsius')
plt.xlabel('temperature_celsius')
plt.ylabel('Demand Multiplier')
plt.show()
```
